Реализовать свой домашний ArrayList
В отличии от стандартного списка, ваш должен быть типизированным.
Для проходного балла - класс должен реализовать протокол Sequence
Для мотивированных и любопытных - класс должен реализовать протокол MutableSequence

Каждый метод из протокола должен быть проверен.


```python
from array import array

class SimpleIterator:
    
    def __iter__(self):
        return self

    def __init__(self, data):
        self.data = data
        self.counter = -1

    def __next__(self):
        if self.counter < len(self.data)-1:
            self.counter += 1
            return self.data[self.counter]
        else:
            raise StopIteration

class ArrayList:
    
    def __init__(self, itemtype, *args):
        print('Calling __init__:')
        self.data = array(itemtype, args)
        print("Initialization: self.__dict__ is: %s" %self.__dict__, end='\n\n\n')
        
    def __getitem__(self, key):
        print('Calling __getitem__:')
        return self.data[key]
    
    def __len__(self):
        print('Calling __len__:')
        return self.data.__len__()
    
    def __contains__(self, item):
        print('Calling __contains__:')
        return item in self.data
    
    def __iter__(self):
        print('Calling __iter__:')
        self.itr = SimpleIterator(self.data)
        return self.itr
    
    def __reversed__(self):
        print('Calling __reversed__:')
        return iter(self.data[::-1])

    def index(self, item):
        print('Calling index:')
        counter = 0
        for element in self.data:
            if element == item:
                return counter
            else:
                counter += 1
        raise ValueError (f"{item} is not in the list. Try again ;)")

    def count(self, item):
        print('Calling count:')
        counter = 0
        for element in self.data:
            if element == item:
                counter += 1
        return counter
        
        
print("Create instance 'al' of 'ArrayList' with type code 'l' and values: 1, 2, 3")
al = ArrayList('l', 1, 2, 3)

# print("ArrayList is: %s" %type(ArrayList), end='\n\n')
# print("ArrayList.__dict__ is: %s" %ArrayList.__dict__, end='\n\n')
# # print("ArrayList methods and properties are: %s" %dir(ArrayList), end='\n\n\n')

# print("al is: %s" %type(al), end='\n\n')
# print("al.__dict__ is: %s" %al.__dict__, end='\n\n')
# print("al methods and properties are: %s" %dir(al), end='\n\n')

# print("al.data is: %s" %al.data, end='\n\n\n')
```
